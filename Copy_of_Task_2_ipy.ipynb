```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a Figure and Axes
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting using Axes object
ax.plot(x, y1, label='Sine wave', color='blue', linewidth=2, linestyle='--')
ax.plot(x, y2, label='Cosine wave', color='red', linewidth=2, linestyle='-.')

# Customizing the Axes object
ax.set_title('Sine and Cosine Waves', fontsize=16)
ax.set_xlabel('X axis label (radians)', fontsize=14)
ax.set_ylabel('Y axis label', fontsize=14)
ax.legend(loc='upper right')
ax.grid(True)

# Display the plot
plt.show()
```

Multiple Subplots OO Api


```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # Create a 2x2 grid of subplots

# Data for each subplot
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

# Top-left subplot
axs[0, 0].plot(x, y1, color='blue', label='Sine')
axs[0, 0].set_title('Sine Wave')
axs[0, 0].grid(True)

# Top-right subplot
axs[0, 1].plot(x, y2, color='red', label='Cosine')
axs[0, 1].set_title('Cosine Wave')
axs[0, 1].grid(True)

# Bottom-left subplot
axs[1, 0].plot(x, y3, color='green', label='Tangent')
axs[1, 0].set_title('Tangent Wave')
axs[1, 0].grid(True)

# Bottom-right subplot
axs[1, 1].plot(x, y4, color='orange', label='Exponential Decay')
axs[1, 1].set_title('Exponential Decay')
axs[1, 1].grid(True)

# Adding padding between subplots
fig.tight_layout()

plt.show()
```

Customizing Plots with Multiple Layers

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Data
x = np.arange(0, 10)
y = np.random.random(10)
z = np.random.random(10)

# Bar plot
ax.bar(x, y, label='Bar plot', color='skyblue', alpha=0.6)

# Line plot
ax.plot(x, z, label='Line plot', color='orange', linewidth=2, marker='o')

# Customizing the plot
ax.set_title('Combined Bar and Line Plot', fontsize=16)
ax.set_xlabel('X axis', fontsize=14)
ax.set_ylabel('Y axis', fontsize=14)
ax.legend()
ax.grid(True)

plt.show()
```

Using Twin Axes (Multiple y-axes)

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.log(x + 1)

# Plotting on the first y-axis
ax1.plot(x, y1, label='Sine wave', color='blue')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Sine wave', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Creating a secondary y-axis
ax2 = ax1.twinx()
ax2.plot(x, y2, label='Logarithmic curve', color='red')
ax2.set_ylabel('Logarithmic curve', color='red')
ax2.tick_params(axis='y', labelcolor='red')

fig.tight_layout()
plt.show()
```

Advanced Customizations

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Customizing the plot
ax.plot(x, y, color='green', linewidth=2, linestyle='-.')

# Customizing ticks
ax.set_xticks([0, np.pi / 2, np.pi, 3 * np.pi / 2, 2 * np.pi])
ax.set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'], fontsize=12, color='purple')

# Customizing the grid
ax.grid(True, which='both', axis='both', linestyle='--', linewidth=0.5)

# Customizing spines (axes lines)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('blue')
ax.spines['left'].set_linewidth(2)

plt.show()
```
